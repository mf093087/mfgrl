```python
import gymnasium as gym
import numpy as np
import random
```

```python
np.random.seed(42)
random.seed(42)
```

```python
env = gym.make("mfgrl:mfgrl/MfgEnv-v0")
obs, info = env.reset(seed=42)
```

```python
total_reward = 0
count = 0
while count <= 5:
    print(f"Step-{count}".center(100, "-"))
    print("cfgs_status: ")
    print(obs["cfgs_status"])
    print()
    print("produced: ")
    print(obs["produced"])

    action = np.random.randint(0, env.action_space.n)
    print(f"Action: {action}")
    
    obs, reward, terminated, truncated, info = env.step(action)
    total_reward += reward
    if terminated or truncated:
        break
    
    count += 1
```

```python
while count <= 20:
    print(f"Step-{count}".center(100, "-"))
    print("cfgs_status: ")
    print(obs["cfgs_status"])
    print()
    print("produced: ")
    print(obs["produced"])

    action = env.action_space.n - 1
    print(f"Action: {action}")
    
    obs, reward, terminated, truncated, info = env.step(action)
    total_reward += reward
    if terminated or truncated:
        break
    
    count += 1
```
