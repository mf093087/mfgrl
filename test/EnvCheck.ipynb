```python
import gymnasium as gym
import numpy as np
import random
```

```python
np.random.seed(42)
random.seed(42)
```

```python
env = gym.make("mfgrl:mfgrl/MfgEnv-v0")
obs, info = env.reset(seed=42)
```

```python
total_reward = 0
count = 0
while count <= 25:
    print(f"Step-{count}".center(100, "-"))
    print("Buffer: ")
    for k, v in env.buffer.items():
        print(f"\t{k}: {v}")
    
    print("\nObs:")
    for k, v in obs.items():
        print(f"\t{k}: {v}")
    
    if count <= 5:
        action = np.random.randint(0, env.action_space.n)
    else:
        action = 5
    print(f"Action: {action}")
    
    obs, reward, terminated, truncated, info = env.step(action)
    total_reward += reward
    if terminated or truncated:
        break
    
    count += 1
```
